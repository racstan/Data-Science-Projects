```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
train_df = pd.read_csv('/content/drive/MyDrive/capabl/train.csv')
test_df = pd.read_csv('/content/drive/MyDrive/capabl/test.csv')
```

```python
print(train_df.describe())
```

```python
train_df = train_df.dropna()
X_train = train_df.drop('trip_duration', axis=1)
y_train = train_df['trip_duration']
```

```python
sample_size = min(10000, len(train_df))
sampled_data = train_df.sample(n=sample_size)

features_to_plot = ['passenger_count', 'pickup_longitude', 'pickup_latitude', 'dropoff_longitude', 'dropoff_latitude']

for feature in features_to_plot:
    plt.figure(figsize=(8, 6))
    sns.scatterplot(x=feature, y='trip_duration', data=sampled_data)
    plt.xlabel(feature)
    plt.ylabel('Trip Duration')
    plt.title(f'Scatter Plot: Trip Duration vs {feature}')
    plt.show()

histogram_sample = sampled_data['trip_duration']

plt.figure(figsize=(8, 6))
sns.histplot(histogram_sample, bins=50, kde=True)
plt.xlabel('Trip Duration')
plt.ylabel('Frequency')
plt.title('Histogram of Trip Duration (Sampled)')
plt.show()
```

```python
model_sample = train_df.sample(n=sample_size)

X_train = model_sample.drop('trip_duration', axis=1)
y_train = model_sample['trip_duration']

X_train_encoded = pd.get_dummies(X_train)

train_encoded_columns = X_train_encoded.columns

linear_model = LinearRegression()
lasso_model = Lasso(alpha=0.1)
ridge_model = Ridge(alpha=1)

linear_model.fit(X_train_encoded, y_train)
lasso_model.fit(X_train_encoded, y_train)
ridge_model.fit(X_train_encoded, y_train)

linear_score = linear_model.score(X_train_encoded, y_train)
lasso_score = lasso_model.score(X_train_encoded, y_train)
ridge_score = ridge_model.score(X_train_encoded, y_train)

print(f"Linear Regression R-squared: {linear_score}")
print(f"Lasso Regression R-squared: {lasso_score}")
print(f"Ridge Regression R-squared: {ridge_score}")
```

```python
test_sample = test_df.sample(n=sample_size)

X_test_encoded = pd.get_dummies(test_sample).reindex(columns=train_encoded_columns, fill_value=0)

y_pred = linear_model.predict(X_test_encoded)

for feature in features_to_plot:
    plt.figure(figsize=(8, 6))
    sns.scatterplot(x=feature, y=y_pred, data=test_sample)
    plt.xlabel(feature)
    plt.ylabel('Predicted Trip Duration')
    plt.title(f'Scatter Plot: Predicted Trip Duration vs {feature} (Test)')
    plt.show()

plt.figure(figsize=(8, 6))
sns.histplot(y_pred, bins=50, kde=True)
plt.xlabel('Predicted Trip Duration')
plt.ylabel('Frequency')
plt.title('Histogram of Predicted Trip Duration for Test Data (Sampled)')
plt.show()

print("Predicted trip durations for the test data:")
print(y_pred)
```
